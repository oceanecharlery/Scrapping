# Scraper EchoJS

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.echojs.com/')
r.status_code
```

```python
soup = BeautifulSoup(r.content)
print(soup.prettify())
```

```python
soup.find_all('a')
```

```python
soup.find_all('a', href=True)
```

```python
links = soup.select('article > h2 > a')

data = []

for link in links:
    data.append({
        'title': link.get_text(),
        'url': link.get('href')
    })

for post in data:
    print('Title: %s' % post['title'])
    print('Url: %s' % post['url'])
    print()
```

# Ecrire un fichier csv

```python
import csv
```

```python
with open('./echojs.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=['title', 'url'])
    writer.writeheader()
    
    for post in data:
        writer.writerow(post)
```

# Scraper YCombinator News

```python
#Titre 
#Nombre de points 
#Nombre de commentaires
```

```python
ycurl = requests.get('https://news.ycombinator.com/')
r.status_code
```

```python
soupyc = BeautifulSoup(ycurl.text)
```

```python
data = []
for tr in soupyc.select('tr.athing'):
    sibling = tr.find_next_sibling()
    data.append({
        'title':tr.find(class_='storylink').string,
        'url':tr.find(class_='storylink').get("href"),
        'user': sibling.find('a').string,
        'score': sibling.find(class_='score').string,
        'comments':sibling.find_all("a")[-1].string
    })
    
for post in data:
    print("Title: %s" % post['title'])
    print("Url: %s" % post['url'])
    print("User: %s" % post['user'])
    print("Score: %s" % post['score'])
    print("Comments: %s" % post['comments'])

    print("")
```

```python
import json
```

```python
with open('./hackernews.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=['title', 'url', 'user','comments','score'])
    writer.writeheader()
    
    for post in data:
        writer.writerow(post)
```

```python
with open('./hackernews.json', 'w') as f:
    json.dump(data,f, ensure_ascii=False, indent=2)
```
